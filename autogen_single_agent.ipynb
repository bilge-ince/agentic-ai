# Single Agent 

This agent develops to calculate and plot Stock Price Change Year to Date for META and Google

```python

llm_config = {
    "config_list": [
        {
            "model": "claude-3-5-sonnet-20240620",
            "api_key": os.environ.get("ANTHROPIC_API_KEY"),
            "api_type": "anthropic",
        }
    ]
}
assistant = AssistantAgent(name="assistant", llm_config=llm_config)
user_proxy = UserProxyAgent(
    name="user_proxy",
    code_execution_config={"work_dir": "code", "use_docker": False},
    human_input_mode="NEVER",
    is_termination_msg=lambda x: x.get("content", "").rstrip().endswith("TERMINATE"),
)
user_proxy.initiate_chat(
    assistant,
    message="Plot a chart of META and GOOG stock price change YTD. Save the result to a file named chart.png.",
)
```
